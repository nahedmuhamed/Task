```python
#Import the needed libararies
import pandas as pd
from google.colab import files
```

```python
#Import the dataframe and print a sample of it
df = pd.read_csv('/content/Employees.csv')
df.head()
```

```python
#Print all the data
print(df)
```

```python
#Understaning information about the data
df.info()
```

```python
#Print summary statistics about numerical columns in the data
df.describe()
```

```python
#Remove duplicates data
new_df = df.drop_duplicates()
print(new_df)
```

```python
#Remove any decimal places in the Age column
new_df['Age'] = new_df['Age'].astype(int)
print(new_df['Age'])
```

```python
#Convert the USD salary to EGP
new_df['Salary(EGP)'] = new_df['Salary(USD)'] * 48.53
new_df.drop('Salary(USD)', axis = 1, inplace = True)
print(new_df['Salary(EGP)'])
```

```python
#Print in the console some stats
print("The Average Age of all Ages = ", new_df['Age'].mean())
print("The Median Salaries = ", new_df['Salary(EGP)'].median())
Ratio = (new_df['Gender'] == 'M').sum() / (new_df['Gender'] == 'F').sum()
print('The Ration between males and female employees = ', Ratio)
```

```python
#Create the new data as CSV file
new_df.to_csv('preprocessed_data.csv',index=False)
print(new_df)
```
